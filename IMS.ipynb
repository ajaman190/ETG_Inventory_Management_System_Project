```python
import json
import time
import secrets
```

```python
file = open('salesRecord.json', 'r')
records = file.read()
file.close()

products = json.loads(records)

file = open('purchaseRecord.json')
records = file.read()
file.close()

purchase = json.loads(records)
```

```python
#function to check product ID already available or not

def checkItem(key):
    product_id = input("Enter the product ID: ")
    
    available = 1
    for i in products:                      # checking whether this ID already is available or not
        if i == product_id:                 # if not then re-enter the product id 
            available = 1
            break
        else:
            available = 0
                
    if available == 1:
        if key == 'A':
            print("\n\nProduct is already available in the store with name:",products[product_id]['name'],"\n---------You need to Update it---------\n\n")

        updateItem(product_id)
        
    else:
        if key == 'D':
            print("\n\nProduct is not available in the store.\n---------You need to Add it---------\n\n")

        addItem(product_id)                     # otherwise add it
        
```

```python
#function to Add product by its ID

def addItem(product_id):
    
    name = input("Enter the product name: ")
    category = input("Enter the product category: ")
    price = int(input("Enter the unit price: "))
    discount = int(input("Enter the discount available: "))
    quantity = int(input("Enter the quantity: "))
    
    products[product_id] = {}
    
    products[product_id]['name'] = name
    products[product_id]['category'] = category
    products[product_id]['price'] = price
    products[product_id]['discount'] = discount
    products[product_id]['quantity'] = quantity
    products[product_id]['stockprice'] = quantity * price
        
    print("Product is successfully added")
```

```python
#function to update product by its ID

def updateItem(product_id):

    name = input("Update the product name: ")
    category = input("Update the product category: ")
    price = int(input("Update the unit price: "))
    discount = int(input("Update the discount available: "))
    quantity = int(input("Update the quantity: "))
       
    products[product_id]['name'] = name
    products[product_id]['category'] = category
    products[product_id]['price'] = price
    products[product_id]['discount'] = discount
    products[product_id]['quantity'] = quantity
    products[product_id]['stockprice'] = quantity * price
        
    print("Product deatils are successfully updated")
    
```

```python
# function for all the sales record or transaction

def salesRecord(sold_item): 
    
    file = open("purchaseRecord.json", 'r')
    txt = file.read()
    file.close()
    
    if len(txt) == 0 :
        file = open("purchaseRecord.json", 'w')
        sold_item = "{"+sold_item+"}"
        file.write(sold_item)
        file.close()
    
    else:
        txt = txt[:-1]
        new = txt + "," + sold_item + "}"
        
        file = open("purchaseRecord.json", 'w')
        file.write(new)
        file.close()
    
```

```python
# Function for buying item from the store


def buyItem():
    
    
    
    """--------------------------------  Buying the product with their ID -----------------------------------"""
    
    hexstr = secrets.token_hex(3)
    customer_id = int(hexstr, 16)
    
    purchased_list = []                       # empty purchased list contains product id and quantity
    buy = 'y'
    available = 1
    
    while (buy == 'y'):
        
        product_id = input("Enter the Product ID: ")
        
        for i in products:                      # checking whether this ID already is available or not
            if i == product_id:                  # if not then re-enter the product id 
                available = 1
                break
            else:
                available = 0
                
        if available == 0:
            print("\n\n-------------- Sorry! Please enter a valid product ID ---------------\n")
            continue
        
        else:
            product_quant = int(input("Enter the Product Quantity: "))

            if (product_quant <= products[product_id]['quantity']): # when required quantity of product is available in stcok


                                                          # updating the product quantity and product stock price

                products[product_id]['quantity'] = products[product_id]['quantity'] - product_quant
                products[product_id]['stockprice'] = products[product_id]['quantity']*products[product_id]['price']
                soldTime = time.ctime()


                sold_item = '\"'+ soldTime +'\":{"customer_id":'+str(customer_id)+',\"product id\":'+product_id+',\"name\":\"'+products[product_id]['name']+'\",\"quantity\":'+ str(product_quant) +',\"total price\":' + str(products[product_id]['price']*(1-(products[product_id]['discount']/100))*product_quant)+'}'

                salesRecord(sold_item)

                                                         # appending the product id and quantity in the purchased list

                purchased_list.append(product_id)
                purchased_list.append(product_quant)
                buy = input('Do you want to buy more item, Press (y/n): ')
                print("\n")


            else:                                                # when required quantity of product is not available in stcok

                print("\n\n------------------------------ Sorry! ------------------------------\n\t",product_id,":",products[product_id]['name'],'quantity available in stock is', products[product_id]['quantity'],"\n")
                buy = input('Do you want to buy more item, Press (y/n): ')
                print("\n")

    
    
    """----------------------------------------  Generating the bill ----------------------------------------"""
    
    count = 1
    price = 0.0
    grand_total = 0.0
    
    print("\n\n\t\t\t\t\t\tAman's Grocery Store\n\n")
    print(time.ctime(),"\t\t\t\t\t\t\t\te-mail: ajaman190@gmail.com")
    
    print("_"*115 + "\n\n")
    
    txt = fixLen("S.No.",15)+fixLen("Product ID",15) + fixLen("Product Name",23) + fixLen("Quantity",12) + fixLen("Unit Price",12) + fixLen("Discount",15) + fixLen("Price After Discount",20)
    print(txt)
    
    for i in range(0, len(purchased_list),2) :
        
        item = purchased_list[i]
        item_quant = purchased_list[i+1]
        
        price = item_quant*(products[item]['price']
                                     - ((products[item]['discount'] * products[item]['price']) /100.0) )
        
        txt = fixLen(str(count),15)+fixLen(item,15)+fixLen(products[item]['name'],23) + fixLen(str(item_quant),12) + fixLen(str(products[item]['price']),12) + fixLen(str(products[item]['discount']),15) + fixLen(str(price),20)
        print(txt)
        
    
        grand_total += price
        count += 1
        
    print("\n\n\t\t\t\t\t\t\t\t\t\tTotal Price : ",grand_total)
    print("_"*115 + "\n") 
    
    print("\n\t\t\t\tNote: Prices are inclusive of all Govt. taxes\n")
    print("\t\t\t\t\tThank You, Please do come again\n")
    
    
    
    
```

```python
def fixLen(str,length):

    text = (str + " "*length)[:length]
    
    return text
```

```python
# Function to print purchase details from the unique customer id

def purchaseDetail():
    
    customer_id = input("Enter the customer id: ")
    
    available = 1
    for i in purchase.keys():                      # checking whether this ID already is available or not
        if str(purchase[i]["customer_id"]) == customer_id:                 # if not then re-enter the product id 
            available = 1
            break
            
        else:
            available = 0
            
    
    if available == 1:
        
        print("Customer ID: ",customer_id)
        
        print("-"*115)
        txt = fixLen("Product ID",20) + fixLen("Product Name",20) + fixLen("Quantity",20) + fixLen("Total Price",20) + fixLen("Purchase date/time",20)
        print(txt)
        
        for i in purchase.keys():
            if str(purchase[i]["customer_id"]) == customer_id :

                txt = fixLen(str(purchase[i]["product id"]),20) + fixLen(purchase[i]["name"],20) + fixLen(str(purchase[i]["quantity"]),20) + fixLen(str(purchase[i]["total price"]),20) + fixLen(i,20)
                print(txt)
                
        print("-"*115)
    
    elif available == 0:
        print("-"*20 + " Invalid Customer ID ! "+ "-"*20)
        
    
```

```python
# Main function

while 1:
    print("\nPress key :\n")
    print("A - To add items in the product list\n")
    print("B - To buy products\n")
    print("C - To show the available product list\n")
    print("D - To update the availble product list\n")
    print("P - To print the purchase details using consumer id\n")
    print("E - To Exit\n")
    
    key = input()

    if key == 'A':
        checkItem(key)
        update()
        file = open('salesRecord.json', 'r')
        records = file.read()
        file.close()

        products = json.loads(records)
        
        file = open('purchaseRecord.json')
        records = file.read()
        file.close()

        purchase = json.loads(records)

        
    elif key == 'B':
        buyItem()
        update()
        file = open('salesRecord.json', 'r')
        records = file.read()
        file.close()

        products = json.loads(records)
        
        file = open('purchaseRecord.json')
        records = file.read()
        file.close()

        purchase = json.loads(records)

        
    elif key == 'C':
        print(products)
        
        
    elif key == 'D':
        checkItem(key)
        update()
        file = open('salesRecord.json', 'r')
        records = file.read()
        file.close()

        products = json.loads(records)
        
        file = open('purchaseRecord.json')
        records = file.read()
        file.close()

        purchase = json.loads(records)

        
    elif key == 'P':
        purchaseDetail()
        
    elif key == 'E':
        print("\n\t\t\t\tThank You! for visiting to our store, Please do come again")
        break
```
